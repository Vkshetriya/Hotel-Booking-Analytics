```python
# convert to datetime
df['check_in_date'] = pd.to_datetime(df['check_in_date'], errors='coerce')
df['check_out_date'] = pd.to_datetime(df['check_out_date'], errors='coerce')

# split datasets
df_clean = df.dropna(subset=['check_in_date', 'check_out_date'])
df_missing = df[df['check_in_date'].isnull()]
```

```python
import re

# function to clean column names
def clean_column(col):
    col = col.lower()                      # lowercase
    col = col.strip()                      # remove spaces
    col = col.replace(" ", "_")            # spaces → underscore
    col = re.sub(r'[^\w_]', '', col)       # remove special characters
    return col

# apply to dataframe
df.columns = [clean_column(col) for col in df.columns]

# check result
df.columns
```

```python
#Convert ALL date columns
date_cols = ['booking_date', 'check_in_date', 'check_out_date', 'travel_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# recreate clean dataframe AFTER fixing dates
df_clean = df.dropna(subset=['check_in_date', 'check_out_date']).copy()
```

```python
# stay duration
df_clean['stay_length'] = (df_clean['check_out_date'] - df_clean['check_in_date']).dt.days

# lead time
df_clean['lead_time'] = (df_clean['check_in_date'] - df_clean['booking_date']).dt.days

# profit
df_clean['profit'] = df_clean['selling_price'] - df_clean['costprice']

# booking month
df_clean['booking_month'] = df_clean['booking_date'].dt.month
```

```python
df['is_cancelled'] = df['booking_status'].str.lower().apply(lambda x: 1 if x == 'cancelled' else 0)
```

## EXPLORATORY DATA ANALYSIS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Booking Distribution by Channel

```python
channel_counts = df['booking_channel'].value_counts()

sns.barplot(x=channel_counts.index, y=channel_counts.values)
plt.xticks(rotation=45)
plt.title("Bookings by Channel")
plt.show()
```

## Interpretation
Certain booking channels generate higher-value bookings, indicating they attract more premium customers.

## Room Type Demand

```python
room_counts = df_clean['room_type'].value_counts()

sns.barplot(x=room_counts.index, y=room_counts.values)
plt.xticks(rotation=45)
plt.title("Room Type Demand")
plt.show()
```

## Interpretation
Premium room types generate higher profit margins, even though their booking volume is lower.

## BOOKING PATTERN ANALYSIS

## Revenue Share by Booking Channel

```python
# aggregate revenue by channel
channel_revenue = df_clean.groupby('booking_channel')['selling_price'].sum()

# plot pie chart
plt.figure(figsize=(7,7))
plt.pie(channel_revenue, labels=channel_revenue.index, autopct='%1.1f%%', startangle=140)

plt.title("Revenue Share by Booking Channel")
plt.show()
```

## Interpretation


## Bookings by Star Rating

```python
sns.countplot(x='star_rating', data=df_clean)
plt.title("Bookings by Star Rating")
plt.show()
```

## Interpretation
4-star hotels have the highest bookings (~9900)
Followed by 3-star hotels (~8600)
5-star hotels have moderate bookings (~3600)
2-star hotels have the lowest bookings (~2500)
Indicates customers prefer mid-range hotels offering value for money

## Room Type vs Star Rating

```python
pivot3 = pd.pivot_table(df_clean,
                        values='customer_id',
                        index='room_type',
                        columns='star_rating',
                        aggfunc='count')

plt.figure(figsize=(8,5))
sns.heatmap(pivot3, annot=True, fmt='.0f', cmap='viridis')

plt.title("Booking Patterns: Room Type vs Star Rating")

plt.show()
```

## Interpretation
Standard rooms in 4-star hotels have the highest bookings (5299)
Standard rooms dominate across all star ratings
Deluxe rooms have moderate demand, highest in 4-star hotels
Suites have the lowest bookings across all star categories
3-star and 4-star hotels show strongest overall demand across room types

## Cancellation Over Time

```python
cancel_trend = df.groupby('booking_date')['is_cancelled'].mean()

plt.figure(figsize=(10,5))
plt.plot(cancel_trend.index, cancel_trend.values)

plt.title("Cancellation Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Cancellation Rate")

plt.show()
```

## Interpreation
Cancellation rate fluctuates between ~10% to 40% over the period
Peak cancellations observed around Jul–Sep 2024
Another moderate spike seen around Dec–Jan
Lower and stable cancellations in Feb–Apr 2025
Overall pattern shows seasonal variation in cancellations
